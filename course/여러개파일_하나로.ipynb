```python
import pandas as pd
import os

#fdir에 경로설정
##경로: 여러개의 엑셀파일이 모아져있는 하나의 파일 경로명
fdir = "/content/drive/MyDrive/2025하이테커/파이썬 강의자료-20251111T001207Z-1-001/파이썬 강의자료/data/대전태평(암반)"
fnames = os.listdir(fdir)

# 폴더 안 파일 목록 가져오기(리스트로)
fnames = os.listdir(fdir)

# 엑셀 파일만 필터링해서 읽기
dfs = []  # 여러 DataFrame을 담을 리스트

for fname in fnames:
    if fname.endswith(".xlsx"):
        filepath = os.path.join(fdir, fname)
        df = pd.read_excel(filepath)
        dfs.append(df)
        print(f"{fname} 불러오기 완료")

# 엑셀 파일 취합하기
combined = pd.concat(dfs, ignore_index=True)
print(combined.head())
```
