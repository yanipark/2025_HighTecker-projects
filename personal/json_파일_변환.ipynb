```python
# module 불러오기
import json
import pandas as pd
import requests
import time


# 사용자 입력 받기
'''
start_date = input("시작 날짜를 입력하세요: ")
end_date = input("끝 날짜를 입력하세요: ")
'''

# api 가져오기
url = "https://apis.data.go.kr/1360000/AsosDalyInfoService/getWthrDataList?serviceKey=8e3214d00f6325655e274255cd6b9472343ae46e38441d7691955c0e17105116&dataType=json&numOfRows=10&pageNo=1&dataCd=ASOS&dateCd=DAY&startDt=20250101&endDt=20251111&stnIds=108"
data = requests.get(url)

time.sleep(1)  # 불러오는 여유시간

print(data.content)
```

```python
print(url)
```

```python
jsondata = json.loads(data.content)
wea_data = jsondata["response"]["body"]["items"]["item"]  #list 가져오기

#리스트를 한눈에 보기 위해
#굳이..,df.copy() 하면 됌
wea_data2 = []

for i in wea_data:
  wea_data2.append(i)
  print(i)
```

```python
w_df = pd.DataFrame(wea_data2)
w_df.head(5)
w_df.isnull().sum()
w_df.describe()
w_df.info()
```

```python
# 필요한 컬럼만 가져오기
w_df = w_df[["stnId","stnNm","tm","avgTa","minTa","sumRn"]]
w_df
```
